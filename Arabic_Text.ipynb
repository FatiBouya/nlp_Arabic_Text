###  Arabic text

```python
 text=["""ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي  وهي بدايات الجبر،ومن المهم فهم كيف كانت هذه الفكرة الجديدة مهمة، فقد كانت خطوة ثورية بعيدا عن
المفهوم اليوناني للرياضيات التي هي في جوهرها  هندسة، الجبركان نظرية موحدة تتحيح الأعداد الكسرية و الأعداد اللا كسرية ، والمقادير الهندسية و غيرها ، أن تتعامل على أنها أجسام جبرية، و أعطت الرياضيات ككل مسارا جديدًا للتطوربمفهوم 
 أوسع بكثير من الذي كان موجودًا من قبل ، وقدم وسيلة للتنمية في هذا الموضوع مستقبلا .و جانب آخر مهم لإدخال أفكار الجبر و هو أنه سمح بتطبيق الرياضيات على نفسها 
بطريقة  لم تحدث من قبل."""]

print(text)
```

### remove ponctuation

```python
import string
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ''' + string.punctuation
text= "".join([char for char in text if char not in punctuations])

text=text.replace("\n", " ")
print(text)
```

### extraire les phrases

```python
from nltk.tokenize import *
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
import nltk
tokenized_text=sent_tokenize(text) 
print(tokenized_text)
```



### extraire les mots

```python
tokenized_text=word_tokenize(text)
print(tokenized_text)
```

### post tagging


```python
tag=nltk.pos_tag(tokenized_text)
print(tag)
```

### chunking with regular Expression

```python
grammar = r"""
  NP: {<DT|PP\$>?<JJ>*<NN>}   # chunk determiner/possessive, adjectives and noun
      {<NNP>+}                # chunk sequences of proper nouns
"""
ns = nltk.RegexpParser(grammar)
l1=[]
for i in nltk.pos_tag(tokenized_text):
  l1.append(i)

print(ns.parse(l1)) 
```

```python
l=[]
#remplaire la liste par les mots du text
for i in tokenized_text:
  l.append(i)

print(l)
```

###  steaming and limmatization 

```python
#stem isn't smart as lemma
ps=PorterStemmer()
#lemma is smart
lemmatizer=WordNetLemmatizer()
for w in l:
 print(w + "::"+ps.stem(w) + "::" +lemmatizer.lemmatize(w))
```
